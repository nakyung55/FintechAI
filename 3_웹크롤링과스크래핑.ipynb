# 개발 방향
- '핀테경제신문'의 최신 기사 제목 수집 (1 Page 기사 전체를 대상으로)
- 저장 위치 : /contents/data

```python
url = "https://fintechtimes.co.kr/news/article_list_all.html"

headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, "html.parser")

soup
```

```python
# 기사 제목 수집

titles = soup.select("ul.art_list_all h2.cmp.c2")
print("수집 건수:", len(titles))

news_list = []

for idx, t in enumerate(titles, 1):
    title_text = t.get_text(strip=True)

    print(f"{idx}. {title_text}")   # 제목 출력

    news_list.append([title_text])
```

```python
df = pd.DataFrame(news_list, columns=["title"])
save_path = "/contents/data"

if not os.path.isdir(save_path):
    os.makedirs(save_path)

df.to_csv(f"{save_path}/news_핀테크경제신문.csv",
          index=False,
          encoding="utf-8-sig")

print("저장 완료:", f"{save_path}/news_핀테크경제신문.csv")
```

# 결과

- HTML 구조 분석 후 ul.art_list_all 영역 내 h2.cmp.c2 태그를 식별해 제목 데이터를 추출
- 총 수집 건수를 확인해 리스트에 저장한 뒤 DataFrame으로 변환
- 수집 결과를 /contents/data/news_핀테크경제신문.csv 경로에 저장
